# RFM

```python
df.dropna(inplace=True)
```

```python
current_date = df["InvoiceDate"].max()
```

```python
#Recency
#Data preprocessing
#df["InvoiceDate"].pd.to_datetime(df["InvoiceDate"])
#calculate recency for each customer
# Calculate Recency
recency_data = df.groupby("CustomerID")["InvoiceDate"].max().reset_index()
recency_data["Recency"] = (current_date - recency_data["InvoiceDate"]).dt.days
recency_data.drop(["InvoiceDate"], axis=1, inplace=True)
recency_data
```

```python
# Calculate Frequency (corrected spelling)
frequency_data = df.groupby("CustomerID")["InvoiceNo"].nunique().reset_index()
frequency_data.columns = ["CustomerID", "Frequency"]
print(frequency_data.head())
```

```python
# Calculate Monetary
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
monetary_data = df.groupby("CustomerID")["TotalPrice"].sum().reset_index()
monetary_data.columns = ["CustomerID", "Monetary"]
print(monetary_data.head())
```

Grouping the RFM

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Merge all data
rfm_data = pd.merge(recency_data, frequency_data, on="CustomerID")
rfm_data = pd.merge(rfm_data, monetary_data, on="CustomerID")
rfm_data
```

```python
#select features for clustering
x=rfm_data[['Recency','Frequency','Monetary']]
x
```

```python
from sklearn.metrics import silhouette_score
```

```python
range_n_clusters = [2, 3, 4, 5, 6]

# Store silhouette scores for different cluster numbers
silhouette_scores = []

# Iterate through different number of clusters and calculate silhouette score
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(x)
    silhouette_avg = silhouette_score(x, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg)

# Choose the optimal number of clusters based on silhouette score
optimal_num_clusters = range_n_clusters[silhouette_scores.index(max(silhouette_scores))]

print("Optimal number of clusters based on silhouette score:", optimal_num_clusters)

# Apply K-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num_clusters, random_state=42)
rfm_data['Cluster'] = kmeans.fit_predict(x)

print(rfm_data.head())
```

# Apriori algorithm

```python
df["Country"].unique()
```

```python
france=df[df["Country"]=="France"].groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
france
```

```python
france_df=pd.DataFrame(france)
france_df
```

```python
france_df["10 COLOUR SPACEBOY PEN"].unique()
```

```python
df["Description"].unique()
```

```python
df["Description"].value_counts()
```

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
    return 0

encoded_basket=france_df.applymap(encode)
encoded_basket
```

```python
#algorithm apply
apr=apriori(encoded_basket,min_support=0.07,use_colnames=True)
```

```python
#creating rule
rules=association_rules(apr)

#sort rules by confidence
rules=rules.sort_values(['confidence'],ascending=True)
rules
```

```python
Australia=df[df["Country"]=="Australia"].groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
Australia
```

```python
Australia["10 COLOUR SPACEBOY PEN"].unique()
```

```python
def encode(x):
  if x>1:
    return 1
  if x<=1:
    return 0

australia_encoded=Australia.applymap(encode)
australia_encoded
```

```python
apr_australia=apriori(australia_encoded,min_support=0.05,use_colnames=True)
```

```python
rules=association_rules(apr_australia)
rules.sort_values(("confidence"),ascending=True)
```

```python
def get_recommendations(product, rules, num_recommendations=5):
    recommendations = []
    for idx, row in rules.iterrows():
        if product in row['antecedents']:
            recommended_item = next(iter(row["consequents"]))
            recommendations.append((recommended_item, row["lift"]))
    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations[:num_recommendations]
```

```python
product_to_recommend = "WHITE HANGING HEART T-LIGHT HOLDER"
recommendations= get_recommendations(product_to_recommend,rules)
print(f"Recommendations for '{product_to_recommend}':")
for item, lift in recommendations:
    print(f"{item} (lift: {lift:.2f})")
```
